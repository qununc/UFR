## 2. Функции

### Функцию можно определить как объект, который принимает аргументы и возвращает значения. Функция определяется с помощью инструкции def. Возврат значений функции реализуется с помощью инструкции return (которая может отсутствовать, в этом случе функция ничего не возвращает).

### С точки зрения работы с функциями, следует различать два понятия - параметр  и аргумент. Список параметров задается в строке определения функции. В свою очередь, аргумент — это реальное значение или ссылка на него, переданное функции при вызове. В Python параметры передаются в функцию по ссылке на объект (call by object reference). Это означает, что при изменении объекта, на который ссылается параметр внутри функции, эти изменения могут отразится и на внешнем по отношению к функции объекте.

### В языке Python проводится четкое разделение между глобальными и локальными переменными (данный вопрос более подробно рассмотривается ниже). Глобальные переменные доступны внтури всего проекта (программы), локальные имеют отношение только к конкретным функциям. Необходимо отметить, что данное деление зависит от контекста.

### Примеры функций

### Пример 1. Расчет факториала

```python
def fact(x):
    fact=1
    for i in range(1,x+1):
        fact=fact*i
    return fact
```

```python
fact(5)
```

### Пример 2. Расчет среднего значения 

### В качестве аргумента функция берет список

```python
def mean(list_num):
    sum=0
    for l in list_num:
        sum+=l
    return sum/len(list_num)   
```

```python
mean(list(range(120)))
```

### Пример 3. Возведение в степень элементов списка

```python
def f_power(list_num,power):
    return {l:l**power for l in list_num}
```

```python
print(f_power(list(range(10)),3))
```

### Рассмотренные выше функции имеют позиционные аргументы (positional). При вызове функций значения в такие аргументы подставляются согласно позиции имён аргументов в определении функции. Передача позиционных аргументов носит обязательный характер. 

### В функцию также могу передаваться так называемые именованные аргументы (named, keyword). В этом случае пердача аргументов не обязательно должна строго следовать порядку, в котором следуют параметры в момент определения функции.

```python
print(f_power(power=4,list_num=list(range(10))))
```

### По правилам языка в момент вызова сначала должны передаваться позиционные аргументы, а затем именованные

```python
print(f_power(power=3,list(range(10))))
```

### В языке Python распространена конструкция вызова функции с параметрами, принимающими некоторое значение по умолчанию.  Их передача носит опциональный харктер. 

```python
def f_power_1(list_num,power=2):
    return {l:l**power for l in list_num}
```

```python
print(f_power_1(list(range(10))))
```

```python
print(f_power_1(list(range(10)),3))
```

### В связи с этим следует учитывать один важный момент - использование знака равенства в момент определения функции и в момент ее вызова носит принципиально разный характер. В пером случае речь идет об объявлении параметров по умолчанию, во втором - о передаче именованных аргументов.

### Количество аргументов, передаваемых в функцию, может вырьировать. Для этих целей существует специальная форма записи с указателем -  * args (args - кортеж значений, * - указатель на место в памяти). Для передачи произвольного количества именованных аргументов используется ** kwargs, в этом случае передается словарь.

```python
def mean(*args):
    sum=0
    for a in args:
        sum+=a
    return sum/len(args)    
```

```python
mean(5,6,7,10,46)
```

```python
mean(5,6)
```

### Порядок следования аргументов не меняется, при этом аргументы, не входящие в кортеж args, должны передаваться как именованные

```python
def f_power_2(*num,power):
    return {n:n**power for n in num}
```

```python
print(f_power_2(7,5,6,9,4,power=2))
```

```python
print(f_power_2(7,5,6,9,power=4))
```

### Функция может быть вложена в другую функцию

```python
def f_nested_square(list_num):
    def f_inner(x):
        return x**2
    return {l:f_inner(l) for l in list_num}
```

```python
print(f_nested_square(list(range(3,69,2))))
```

### Функция может возвращать несколько значений (фактически возвращается кортеж)

### Пример 4. Расчет нескольких статистических показателей

```python
def statistics(list_num):
    sum=0
    for l in list_num:
        sum+=l
    return sum, sum/len(list_num) 
```

### Анонимные функции 

### Во многих языках программирования, в т.ч. и в языке Python, существует специальный вид функций - анонимные функции. Анонимные функции (т.е. функции, не имеющие имени) следуют упрощенным (но и более) строгим синтаксическим правилам. Данные функции могут определяться в любом месте программы с помощью ключевого слова lambda.

### Анонимные функции обычно используются как вспомогательные внутри функций более высокого порядка,  т.е. функций, которые берут в качестве аргументов другие функции. Например, они широко применяются для работы с объектами dataframe библиотеки pandas внутри конструкции apply.

## 3. Объекты-генераторы

### Генераторы — функции, которые можно приостанавливать и возобновлять во время их выполнения. Они возвращают объект, который можно итерировать. В отличие от списков, генераторы ленивы (lazy) и поэтому работают с текущим элемент только по запросу.  Таким образом, они намного эффективнее используют память при работе с большими наборами данных.  С точки зрения синтаксиса, от обычных функций они отличаются тем, что вместо оператора return используется оператор yield.

### Пример генератора, который возводит в заданную степень последовательность чисел

```python
def g_power(n,power): 
    for i in range(n):## функция range, повсеместно используемая в Python, является примером генератора
        yield  i ** power
```

```python
for i in g_power(5,3): 
    print(i, end=' : ')
```

```python
x=g_power(5,3)
```

```python
next(x)
```

```python
next(x)
```

```python
list(g_power(8,2))
```

### Вариант реализации с циклом while

```python
def g_power_1(power): 
    i = 1  
    while True: 
        print(i)
        yield i**power                
        i += 1 
```

```python
x=g_power_1(3)
```

```python
next(x)
```

```python
next(x)
```

```python
next(x)
```

```python
for num in g_power_1(4): 
    if num > 100: 
         break    
    print(num) 
```

```python
for i in even_gen:
    print(i)
```

## 4. Пространство имен

### Пространство имён (namespace) — это набор связей имён с объектами. Python реализует пространства имен в виде словарей.

### Такие понятия как пространство имен (namespace) и область видимости переменных (scope) имеют большое значение в процессе разработки программ,состоящих их большого количества модулей и функций. Глобальное пространство имен - пространство имен, которое включает имена из различных импортированных модулей. Оно создается в момент импорта модуля в проект и существует до конца выполнения скрипта. Локальное пространство имен включает локальные имена внутри функции.  Это пространство имен создается при вызове функции и существует до тех пор, пока функция не вернет результат.

### Пространство встроенных имен (модуль builtins загружается по умолчанию)

```python
print(dir(__builtins__))
```

### Таблица глобального пространство имен вызывается с помощью функции globals(). Для вызова локального пространства имен служит функция locals(). В глобальной среде проекта globals() и locals() будут совпадать.

```python
print(globals())
```

### Локальное пространство имен существует внутри функции

```python
def sum_(x,y):
    z=3
    print('Local objects',locals())
    return x**z+y
```

```python
sum_(5,6)
```
